```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the dataset
dataset = pd.read_csv("Store.csv")
dataset.head()
```

```python
df = dataset.copy()
df = df.groupby('OrderID')['ProductName'].apply(list).reset_index() #for each OrderID create list of all ProductNames
df = pd.DataFrame(df['ProductName'].to_list())  #Extend each value in a list to its own column

df.head()
```

```python
df.to_csv('Store.csv', index=False)
```

```python
# apriori wants this dataset to be in a list of lists
rows = len(df)
columns = len(df.columns)

transactions_list = []
for i in range(0, rows):
    transactions_list.append([str(df.values[i, j]) for j in range(0, columns) if df.values[i, j] is not None])
    #loops through each row (i) in the df, then another loop to grab the value for each column (j), and appends the value into the list
transactions_list[:3]
```

```python
#Train the model

from apyori import apriori  #from apyori library we are calling the apriori function

model = apriori(transactions = transactions_list, 
                min_support = 0.05, #why because usually the support is between 0.01 to 0.05 with 0.05 having fewer items but more realiable 
                min_confidence = 0.2, #why 0.20 because I want to see items from 100% to 20% confidence
                min_lift = 3,         # a good lift in general is 3
                min_length = 2,       # a minimum of two products
                max_length = 2)       # a maximum of two products 
model
```

```python
results = list(model)
#results
```

```python
def inspect(results):
    lhs =        [tuple(x[2][0][0])[0] for x in results]
    rhs =        [tuple(x[2][0][1])[0] for x in results]
    supports =   [x[1] for x in results]
    confidences= [x[2][0][2] for x in results]
    lifts =      [x[2][0][3] for x in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))

results_df = pd.DataFrame(inspect(results), columns=['Left Hand Side','Right Hand Side','Support','Confidence','Lifts'])
results_df[['Support', 'Confidence']] = results_df[['Support', 'Confidence']] * 100
results_df.sort_values(by='Lifts', ascending=False)
```

```python
#The top three roles are:
#- There is an 86% chance that when customers purchase Children's Cutlery Dolly Girl, they will also purchase Children's Cutlery Spaceboy, which appears in 5.67% of the transactions and has the highest Lift of 14.08. The Lift says that when the Dolly Girl is purchased, it is 14.08 times likely that the Spaceboy will be purchased as well. 
#- There is a 79% chance that when the customer purchases an Alarm Clock Bakelike Green, they will also purchase the Red one, which appears in 7.03% of the transactions and has the second highest Lift of 9.47. 
#- There is a 72.5% chance that when the customer purchases an Alarm Clock Bakelike Pink, they will also purchase the Red one, which appears in 6.57% of the transactions and has the third highest Lift.
```
