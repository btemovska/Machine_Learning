```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
#each row represents a customer and what they bought
dataset
```

##### Data Preprocessing

```python
transactions_list = []
for _, row in dataset.iterrows():
    transaction = [item for item in row if str(item) != 'nan']
    if transaction:  # Only append non-empty transactions
        transactions_list.append(transaction)

transactions_list[:3]
```

##### Training the Eclat model

```python
from apyori import apriori
```

```python
model = apriori(transactions = transactions_list, 
                min_support = 0.003, 
                min_confidence = 0.2, 
                min_lift = 3,
                min_length = 2, 
                max_length = 2)
model
```

```python
results = list(model)
#results
```

```python
def inspect(results):
    lhs =        [tuple(x[2][0][0])[0] for x in results]
    rhs =        [tuple(x[2][0][1])[0] for x in results]
    supports =   [x[1] for x in results]
    return list(zip(lhs, rhs, supports))

results_df = pd.DataFrame(inspect(results), columns=['Product1','Product2','Support'])
results_df.sort_values(by='Support', ascending=False)
```
