```python
# Import Libraries
import pandas as pd
import numpy as np
```

```python
# Read the dataset
dataset = pd.read_csv("Store.csv")
dataset.sample(5)
```

```python
df = dataset.copy()
df = df.groupby('OrderID')['ProductName'].apply(list).reset_index() #for each OrderID create list of all ProductNames
df = pd.DataFrame(df['ProductName'].to_list())  #Extend each value in a list to its own column
df.sample(5)
```

#### Create a List of Lists to prepate for encoding 

```python
# apriori wants this dataset to be in a list of lists
rows = len(df)
columns = len(df.columns)

transactions_list = []
for i in range(0, rows):
    transactions_list.append([str(df.values[i, j]) for j in range(0, columns) if df.values[i, j] is not None])
    #loops through each row (i) in the df, then another loop to grab the value for each column (j), and appends the value into the list
transactions_list[:3]
```

#### Transactionalize it

```python
from mlxtend.preprocessing import TransactionEncoder

TE = TransactionEncoder()
array = TE.fit(transactions_list).transform(transactions_list)

cleaned_df = pd.DataFrame(array, columns=TE.columns_)
cleaned_df.sample(5)
```

```python
dataset['ProductName'].nunique()  #number of columns on cleaned_df matches the number of unique product values
```

#### Train the model

```python
from mlxtend.frequent_patterns import apriori, association_rules

model = apriori(cleaned_df, min_support = 0.05, use_colnames=True)
model.head()
```

```python
results = association_rules(model, metric='lift', min_threshold=1, num_itemsets=len(model))
results = results.sort_values(by='lift', ascending=False)
results = results[['antecedents', 'consequents', 'support', 'confidence', 'lift', 'representativity']] #there are ore columns
results
```
