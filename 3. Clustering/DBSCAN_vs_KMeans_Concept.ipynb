#### Import library

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:95% !important; }</style>"))
```

#### Import dataset

```python
blobs = pd.read_csv('cluster_blobs.csv')
moons = pd.read_csv('cluster_moons.csv')
circles = pd.read_csv('cluster_circles.csv')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))

sns.scatterplot(data=blobs, x='X1', y='X2', ax=axes[0])
axes[0].set_title('Blobs')

sns.scatterplot(data=moons, x='X1', y='X2', ax=axes[1])
axes[1].set_title('Moons')

sns.scatterplot(data=circles, x='X1', y='X2', ax=axes[2])
axes[1].set_title('Circles')

plt.tight_layout()
plt.show()
```

#### FIT Model

```python
from sklearn.cluster import DBSCAN
model = DBSCAN()
blobs_labels = model.fit_predict(blobs)

model = DBSCAN(eps=0.10)
moons_labels = model.fit_predict(moons)

model = DBSCAN(eps=0.10)
circles_labels = model.fit_predict(circles)


#Visualize them
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
sns.scatterplot(data=blobs, x='X1', y='X2', hue=blobs_labels, palette = 'Set1', ax=axes[0])
sns.scatterplot(data=moons, x='X1', y='X2', hue=moons_labels, palette = 'Set1', ax=axes[1])
sns.scatterplot(data=circles, x='X1', y='X2', hue=circles_labels, palette = 'Set1', ax=axes[2])
plt.tight_layout()
plt.show()

#Blobs three clusters and the red are outliers (clustered as -1)
```

#### Comparing to KMeans

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)

blobs_labels = model.fit_predict(blobs)
moons_labels = model.fit_predict(moons)
circles_labels = model.fit_predict(circles)

#Visualize them
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
sns.scatterplot(data=blobs, x='X1', y='X2', hue=blobs_labels, palette = 'Set1', ax=axes[0])
sns.scatterplot(data=moons, x='X1', y='X2', hue=moons_labels, palette = 'Set1', ax=axes[1])
sns.scatterplot(data=circles, x='X1', y='X2', hue=circles_labels, palette = 'Set1', ax=axes[2])
plt.tight_layout()
plt.show()
```
